```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
bank_df = pd.read_excel('/content/drive/MyDrive/ASAC_최종프로젝트/data/금융감독원데이터/(저축)은행_기초재무정보주요경영지표_230630기준.xlsx')
```

```python
bank_df
```

```python
bank_df['총자산대비부채'] = bank_df['총부채'] / bank_df['총자산']
bank_df['총자산대비자기자본'] = bank_df['자기자본'] / bank_df['총자산']
```

```python
bank_df['BIS비율_score'] = pd.qcut(bank_df['BIS비율'], q=10, labels=range(1, 11, 1))
bank_df['여신비율_score'] = pd.qcut(bank_df['여신비율'], q=10, labels=range(10, 0, -1))
bank_df['ROA_score'] = pd.qcut(bank_df['ROA'], q=10, labels=range(1, 11, 1))
bank_df['총자산대비부채_score'] = pd.qcut(bank_df['총자산대비부채'], q=10, labels=range(10, 0, -1))
bank_df['총자산대비자기자본_score'] = pd.qcut(bank_df['총자산대비자기자본'], q=10, labels=range(1, 11, 1))
```

```python
bank_df['total_score'] = bank_df['BIS비율_score'].astype(float) + bank_df['여신비율_score'].astype(float) + bank_df['ROA_score'].astype(float) + bank_df['총자산대비부채_score'].astype(float) + bank_df['총자산대비자기자본_score'].astype(float)
```

```python
bank_df['안정성등급'] = pd.qcut(bank_df['total_score'], q=3, labels=range(1, 4))
```
