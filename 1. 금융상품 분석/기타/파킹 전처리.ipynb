```python
grade_dict = {
    'OSB저축은행': '오에스비저축은행',
    'JT친애저축은행': '제이티친애저축은행',
    'OK저축은행': '오케이저축은행',
    'DGB대구은행':'대구은행',
     'DB저축은행' :'디비저축은행',
    '엠에스저축은행': '엠에스상호저축은행',
    'DH저축은행': '디에이치저축은행',
    'JT저축은행': '제이티저축은행',
    'BNK저축은행': '비엔케이저축은행',
    '푸른저축은행': '푸른상호저축은행',
    'SBI저축은행': '에스비아이저축은행',
    '유니온저축은행': '유니온상호저축은행',
    'IBK저축은행': '아이비케이저축은행',
    '세람저축은행': '세람상호저축은행',
    'KB저축은행': '케이비저축은행',
    '대원저축은행': '대원상호저축은행',
    '카카오뱅크' : '주식회사 카카오뱅크',
    'KB국민은행' : '국민은행',
    'IBK기업은행' : '중소기업은행',
    '토스뱅크' : '토스뱅크 주식회사',
    'NH농협은행' : '농협은행주식회사',
    '케이뱅크' : '주식회사 케이뱅크',
    '평택저축은행' : '평택상호저축은행',
    'KDB산업은행' : '한국산업은행',
    'SNT저축은행' : '에스앤티저축은행',
    'SC제일은행' : '한국스탠다드차타드은행',
    'SH수협은행' : '수협은행',
    'NH저축은행' : '엔에이치저축은행'
}
```

```python
def bank_grade(df):
    df['금융회사'] = df['금융회사'].apply(lambda x: grade_dict[x]if x in grade_dict.keys() else x )
    df = pd.merge(left = df, right = bank_info_df[['구분','등급']], how = 'left',
            left_on = '금융회사' , right_on = "구분")
    df.drop(columns = '구분', inplace = True)
    return df
```

```python
# NaN값 우대사항 없음으로 처리
df['우대사항정리'].fillna("우대사항없음", inplace = True)
```

```python
# 대분류 확인하기
cate_set = set()

for i in range(len(df)):
    text = df.loc[i,'우대사항정리' ]
    text_list = text.split("\n")
    for t in text_list :
    # 대분류
        cate = t.split(":")[0].replace("*","").strip()
        cate_set.add(cate)
cate_set
```

```python
# 다른 값 통일시켜주기 - 마케팅 동의
df['우대사항정리'] = df['우대사항정리'].apply(lambda x : x.replace("마케팅 동의", "마케팅동의") if "마케팅 동의" in x else x)
```

```python
df.columns
```

```python
col_list_all_prefer = ['급여이체','마케팅동의','자동이체','카드실적','최초거래','주택청약',
                '비대면','영업점','해당은행','이체횟수','만기','변동금리','이벤트','기타',
                '나이','환경','신혼','반려인','직장인','여행','금액구간']
col_list_all = ['금융회사', '상품명', '구간', '세전이자율', '세후이자율', '세후이자(예시)', '최고우대금리', '실질이자율',
       '비교 공시일', '담당부서 및 연락처', '상품유형', '우대조건', '가입대상', '가입방법', '금액', '이자지급',
       '기타유의사항', '예금자\n보호', '유형','등급', '우대사항정리'] + col_list_all_prefer
```

```python
# 모든 리스트에 따라 df의 col생성
for c in col_list_all_prefer:
    df[c] = np.nan
```

```python
# 금리 추출
p = re.compile("[0-9]+[.][0-9]+[%]")


for i in range(len(df)):
    text = df.loc[i,'우대사항정리' ]
    if text == "우대사항없음":
        continue
    text_list = text.split("\n") # 우대사항 text 가지고오기
    for t in text_list :
        # 카테고리
        if len(t) <3 :
            continue
        cate = t.split(":")[0].replace("*","").strip()
        # 금리
        try:
            rate = p.findall(t.split(":")[1])[0]
        except:
            p2 = re.compile("[0-9][%]")
            rate = p2.findall(t.split(":")[1])
        df.loc[i,cate ] = rate
```

```python
# 은행 안정성 합치기
df = bank_grade(df)
```
